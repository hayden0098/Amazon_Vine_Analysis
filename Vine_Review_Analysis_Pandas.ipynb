```python
# import the pandas dependency
import pandas as pd
```

```python
data_to_load = 'vine_table.csv'
vine_df = pd.read_csv(data_to_load)
vine_df
```

```python
# The data is filtered to create a DataFrame or table where there are 20 or more total votes 
moreThan20_total_votes_df = vine_df[vine_df["total_votes"] >= 20]
moreThan20_total_votes_df
```

```python
# The data is filtered to create a DataFrame or table where the percentage of helpful_votes 
#is equal to or greater than 50% 
greaterThan50_votes_percentage_df = moreThan20_total_votes_df[(moreThan20_total_votes_df["helpful_votes"]/vine_df["total_votes"]) >= 0.5]
greaterThan50_votes_percentage_df
```

```python
# The data is filtered to create a DataFrame or table where there is a Vine review
paid_vine_program_df = greaterThan50_votes_percentage_df[greaterThan50_votes_percentage_df["vine"] == "Y"]
total_paid_vine = paid_vine_program_df.count()
paid_vine_program_df
```

```python
# The data is filtered to create a DataFrame or table where there isn’t a Vine review
unpaid_vine_program_df = greaterThan50_votes_percentage_df[greaterThan50_votes_percentage_df["vine"] == "N"]
total_unpaid_vine = unpaid_vine_program_df.count()
unpaid_vine_program_df
```

```python
# The total number of reviews
total_reviews = greaterThan50_votes_percentage_df["review_id"].count()
total_reviews
```

```python
# The total number of 5-star reviews
total_5star_reviews = greaterThan50_votes_percentage_df[greaterThan50_votes_percentage_df["star_rating"] == 5].count()
total_5star_reviews
```

```python
#the percentage 5-star reviews are calculated for all Vine reviews
paid_5star_reviews = paid_vine_program_df[paid_vine_program_df['star_rating'] == 5].count()
paid_percentage_5star_reviews = round(paid_5star_reviews / total_paid_vine * 100,2)
paid_5star_reviews
```
